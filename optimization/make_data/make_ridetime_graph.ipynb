```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import os

def calculate_time(minutes):
    start_time = 9 * 60
    total_minutes = start_time + minutes
    hours = total_minutes // 60
    mins = total_minutes % 60
    if hours >= 24:
        hours -= 24
    return f"{hours:02d}:{mins:02d}"

# CSV
df = pd.read_csv("../attraction_heatmap5.csv", index_col=0)

# x
x_values = list(map(int, df.columns))
filtered_indices = [i for i, x in enumerate(x_values) if x <= 676]
x_values = [x_values[i] for i in filtered_indices]
x_labels = [calculate_time(x) for x in x_values]

# 15
x_ticks = [x for x in x_values if x % 15 == 0]
x_tick_labels = [calculate_time(x) for x in x_ticks]

# 
df = df.iloc[:, filtered_indices]

# 
colors = cm.get_cmap('tab20', len(df))

# 
output_folder = "ridetime_graph"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 
for idx, (name, row) in enumerate(df.iterrows()):
    plt.figure(figsize=(20, 6))  # 
    plt.bar(x_values, row, color=colors(idx), alpha=0.8, width=10)  # 
    plt.title(f"{name}", fontsize=16)
    plt.grid(True, which="both", linestyle="--", linewidth=1)
    plt.ylabel("", fontsize=14)
    plt.xlabel("", fontsize=14)
    plt.xticks(ticks=x_ticks, labels=x_tick_labels, rotation=45, ha='right', fontsize=10)
    plt.tight_layout()

    # 
    file_path = os.path.join(output_folder, f"{name}_ridetime_graph.png")
    plt.savefig(file_path)
    plt.close()
    print(file_path,"")
```
