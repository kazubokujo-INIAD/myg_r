### いつの時間帯にどのアトラクションを乗るのか

```python
import pandas as pd
import os

attractions = {
    0: 'ソアリン',
    1: 'トイストーリーマニア',
    2: 'タワーオブテラー',
    3: 'センターオブジアース',
    4: 'インディージョーンズ',
    5: 'レイジングスピリッツ',
    6: 'マジックランプシアター',
    7: 'タートルトーク'
}

# CSVファイルを読み込む
df = pd.read_csv('../updated_combined_data2.csv')

# '最適プラン'と'タイムスケジュール'を取り出し、1次元配列に変換
optimal_plan = [eval(plan) for plan in df["最適プラン"]]
time_schedule = [eval(plan) for plan in df["タイムスケジュール"]]
#print(optimal_plan)
#print(time_schedule)

# 新しいデータフレームを作成（初期値を0で埋める）
result_df = pd.DataFrame(data=0, index=attractions.values(), columns=list(range(0, 721, 5)))
#display(result_df)

# '最適プラン'と'タイムスケジュール'を対応させて分別
c = 0
for y in optimal_plan:
    x = time_schedule[c]
    for i in range(len(y)):
        index = attractions.get(y[i])
        #print(f"Attraction: {index}, Time: {x[i]}")
        
        # .loc[]を使って結果を更新
        result_df.loc[index, x[i]] += 1
        #display(result_df)
    
    c += 1

result_df.to_csv(os.path.join('attraction_heatmap2.csv'), index=True)
```
