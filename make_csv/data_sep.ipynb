```python


import pandas as pd
import os
from datetime import datetime, timedelta

# 入力フォルダと出力フォルダ
input_folder = 'folder'
output_folder = 'split_output'
os.makedirs(output_folder, exist_ok=True)

# フォルダ内のCSVファイルを処理
for filename in os.listdir(input_folder):
    if filename.endswith('.csv'):
        file_path = os.path.join(input_folder, filename)

        # ファイル読み込み
        df = pd.read_csv(file_path)

        # 日付をファイル名から抽出（例: DisneySea_2024_9_waittime.csv → 2024-09-01）
        parts = filename.split('_')
        year = int(parts[1])
        month = int(parts[2])
        start_date = datetime(year, month, 1)

        # 行数と分割数
        total_rows = df.shape[0]
        rows_per_day = 48
        num_days = total_rows // rows_per_day

        # 分割して保存
        for i in range(num_days):
            day_df = df.iloc[i * rows_per_day : (i + 1) * rows_per_day].copy()
            current_date = start_date + timedelta(days=i)
            output_filename = current_date.strftime('%Y-%m-%d') + '.csv'
            output_path = os.path.join(output_folder, output_filename)
            day_df.to_csv(output_path, index=False)
```
