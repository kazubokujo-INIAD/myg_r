```python
import pandas as pd
import numpy as np
```

```python
TDS_22_7=pd.read_csv('data/TDS_22_7.csv',header=0)
TDS_22_8=pd.read_csv('data/TDS_22_8.csv',header=0)
TDS_23_7=pd.read_csv('data/TDS_23_7.csv',header=0)
TDS_23_8=pd.read_csv('data/TDS_23_8.csv',header=0)
```

```python
print(TDS_22_7.shape)
print(TDS_22_8.shape)
print(TDS_23_7.shape)
print(TDS_23_8.shape)
```

```python
display(TDS_22_7.head())
display(TDS_22_8.head())
display(TDS_23_7.head())
display(TDS_23_8.head())
```

```python
import pandas as pd

# 縦連結
merged_df = pd.concat([TDS_22_7, TDS_23_7], ignore_index=True)
merged_df2 = pd.concat([TDS_22_8, TDS_23_8], ignore_index=True)

# 結果を新しいCSVファイルとして保存
output_file = 'TDS_7.csv'
output_file2 = 'TDS_8.csv'
merged_df.to_csv(output_file, index=False)
merged_df2.to_csv(output_file2, index=False)

print(f"結合されたCSVファイルが {output_file} に保存されました。")
print(f"結合されたCSVファイルが {output_file2} に保存されました。")
```
