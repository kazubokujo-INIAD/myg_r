```python
import pandas as pd
import os

# フォルダパスを指定してください
folder_path = '../graduation-research/date_data2/'

# 初期化：データフレームを格納するリスト
data_frames = []
file_count = 0  # ファイル数をカウント

# フォルダ内のすべてのファイルを処理
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        
        # CSVファイルを読み込み、一列目をインデックスとして設定
        df = pd.read_csv(file_path, index_col=0)
        
        # 行と列を入れ替える（転置）
        transposed_df = df.T
        
        # データフレームをリストに追加
        data_frames.append(transposed_df)
        
        # ファイル数をカウント
        file_count += 1

# 平均を計算
if file_count > 0:
    # 平均を計算
    average_df = sum(data_frames) / file_count

    # 合計結果を表示
    print("Average DataFrame:")
    display(average_df)

    # 合計結果をCSVファイルとして保存
    output_file_path = os.path.join(folder_path, 'average_data.csv')
    average_df.to_csv(output_file_path)
    print(f"\n平均データフレームを {output_file_path} に保存しました。")

    # 中央値を計算
    median_df = pd.concat(data_frames).groupby(level=0).median()

    # 中央値結果を表示
    print("Median DataFrame:")
    display(median_df)

    # 中央値結果をCSVファイルとして保存
    median_output_file_path = os.path.join(folder_path, 'median_data.csv')
    median_df.to_csv(median_output_file_path)
    print(f"\n中央値データフレームを {median_output_file_path} に保存しました。")

    # 最頻値を計算
    mode_df = pd.concat(data_frames).groupby(level=0).agg(lambda x: x.mode().iloc[0])

    # 最頻値結果を表示
    print("Mode DataFrame:")
    display(mode_df)

    # 最頻値結果をCSVファイルとして保存
    mode_output_file_path = os.path.join(folder_path, 'mode_data.csv')
    mode_df.to_csv(mode_output_file_path)
    print(f"\n最頻値データフレームを {mode_output_file_path} に保存しました。")
else:
    print("CSVファイルが見つかりませんでした。")
```

```python
import pandas as pd

def remove_rows_without_labels(csv_file_path, output_file_path):
    # CSVファイルを読み込む
    df = pd.read_csv(csv_file_path)

    # 列ラベルがない行を削除（すべての列がNaNの場合）
    df_cleaned = df.dropna(how='all').reset_index(drop=True)

    # 新しいCSVファイルとして保存
    df_cleaned.to_csv(output_file_path, index=False)
    print(f"列ラベルのない行を削除したデータフレームを '{output_file_path}' に保存しました。")

# 使用例
input_csv_path = 'rre_mode_data.csv'  # 入力CSVファイルのパス
output_csv_path = 'rrr_mode_data.csv'  # 出力CSVファイルのパス

remove_rows_without_labels(input_csv_path, output_csv_path)
```
